```python
# imporitng the pandas package as pd

import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
Movie =pd.read_csv("movie.txt", header=None, names=[ 'ID','Name_of_Movie','Year','Ratings','Duration' ])
```

### Q1. Number of Movies Released in a year between 1990 to 2000

```python
# TO FIND THE NUMBER OF MOVIES BETWEEN YEAR 1990 to 2000 =
# year1 is a temporary you can overwrite it.

year1= Movie[(Movie['Year']>=1990) & (Movie['Year']<=2000)]
year1.drop(['ID','Ratings','Duration'],inplace=True, axis=1)
year1.count()
```

### Q2. To find the number of Movies whose ratings are more than 4.

```python
# Ratings of movies more than 4=

Movie[Movie['Ratings']>4] 
```

### Q3. TO Find the movies whose ratings are between 2 and 4.

```python
#  Movie (ratings) greater than 2 and less then 4. If you just want the count of the movies with ratings put ".count()"in front of R 
# R is just temporary dataframe you can over-write it.

R= Movie[(Movie['Ratings']>=2) & (Movie['Ratings']<=4)]
R.drop(['ID','Year','Duration'],inplace=True, axis=1)
R
```

### Q4. Find the number of movies with duration more than 2 hours (7200 second).

```python
# To Find the Movies having duration more than 2hours 7200 seconds
# If you only want the count store it in a variable and print.

Movie[Movie[4] > 7200]
```
